```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv(r'E:\Data Science\Projects\ML\walmart-recruiting-store-sales-forecasting\walmart-recruiting-store-sales-forecasting-data\train.csv')
train.head()
```

```python
test = pd.read_csv(r'E:\Data Science\Projects\ML\walmart-recruiting-store-sales-forecasting\walmart-recruiting-store-sales-forecasting-data\test.csv')
test.head()
```

```python
stores = pd.read_csv(r'E:\Data Science\Projects\ML\walmart-recruiting-store-sales-forecasting\walmart-recruiting-store-sales-forecasting-data\stores.csv')
stores.head()
```

```python
features = pd.read_csv(r'E:\Data Science\Projects\ML\walmart-recruiting-store-sales-forecasting\walmart-recruiting-store-sales-forecasting-data\features.csv')
features.head()
```

```python
sampleSubmission = pd.read_csv(r'E:\Data Science\Projects\ML\walmart-recruiting-store-sales-forecasting\walmart-recruiting-store-sales-forecasting-data\sampleSubmission.csv')
sampleSubmission.head()
```

```python
feat_sto = features.merge(stores, on = 'Store')
feat_sto.head()
```

```python
feat_sto.dtypes
```

```python
feat_sto.Date = pd.to_datetime(feat_sto.Date)
train.Date = pd.to_datetime(train.Date)
test.Date = pd.to_datetime(test.Date)
```

```python
feat_sto['Week'] = feat_sto.Date.dt.week 
feat_sto['Year'] = feat_sto.Date.dt.year
```

```python
train_detail = train.merge(feat_sto, on = ['Store','Date','IsHoliday']).sort_values(by = ['Store', 'Dept', 'Date']).reset_index(drop = True)
```

```python
test_detail = test.merge(feat_sto, on = ['Store','Date','IsHoliday']).sort_values(by = ['Store', 'Dept', 'Date']).reset_index(drop = True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
train_detail['Type'] = le.fit_transform(train_detail['Type'])
test_detail['Type'] = le.fit_transform(test_detail['Type'])
```

```python
train_detail
```

```python
train_detail = train_detail.fillna(0)
test_detail = test_detail.fillna(0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_jobs = 4)
```

```python
X = train_detail.drop(['Weekly_Sales', 'Date'], axis = 1)
y = train_detail['Weekly_Sales']
X_test = test_detail.drop(['Date'], axis = 1)
```

```python
rf.fit(X, y)
```

```python
result = rf.predict(X_test)
```

```python
sampleSubmission["Weekly_Sales"] = result
```
